```python
# Crear un DataFrame de ejemplo
data = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.rand(100) * 10,
    'C': np.random.randint(1, 5, 100),
    'Category': np.random.choice(['Group 1', 'Group 2'], 100)
})

# Generar violin plots subdivididos por una variable categórica, con colores personalizados y líneas de referencia
eda.violin_plots(
    data,
    columns=['A', 'B'],
    category='Category',
    palette="Set2",
    y_reference_lines=[0, 5]
)
```

```python

# Generar boxplots subdivididos por una variable categórica, con colores personalizados y líneas de referencia
eda.box_plots(
    data,
    columns=['A', 'B'],
    category='Category',
    palette="Set1",
    y_reference_lines=[-2, 2]
)
```
